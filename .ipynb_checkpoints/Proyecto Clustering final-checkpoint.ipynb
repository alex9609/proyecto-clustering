## Analisis de distribución

```python
int_cols = df_delete.select_dtypes(exclude='object').columns
int_cols #Tomamos los valores enteros
```

Analizamos la distribución de los datos con boxplot


```python
def boxplot(df,cols):
    for _ in cols:
        plt.figure(figsize=(7,7))
        sns.boxplot(data = df,y = _)
        plt.show()
```

```python
boxplot(df_delete,int_cols)
```

Quitando outliers de edad

```python
df_delete = df_delete[['ID', 'Year_Birth', 'Income', 'Kidhome', 'Teenhome', 'Recency',
       'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts',
       'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases',
       'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases',
       'NumWebVisitsMonth']]
df_delete
```

```python
df_outliers = df_delete[(df_delete['Income'] < 600000) & (df_delete['Year_Birth'] > 1920) &
                          (df_delete['MntMeatProducts'] < 1500)  & (df_delete['MntSweetProducts'] < 250) &
                          (df_delete['MntGoldProds'] < 255)  & (df_delete['NumDealsPurchases'] < 14) &
                          (df_delete['NumWebPurchases'] < 20)  & (df_delete['NumCatalogPurchases'] < 20) &
                          (df_delete['NumWebVisitsMonth'] < 17.5)]
df_outliers
```

```python
boxplot(df_outliers,df_outliers.columns)
```

## Escalar los datos

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_customer_scaled = scaler.fit_transform(df_outliers.drop('ID',axis=1)) #Escalamos los datos eliminando la columna ID que no nos brinda mayor informacion
df_customer_scaled
```

## PCA para clustering

Analisis de componentes principales

```python
df_customer_scaled = pd.DataFrame(df_customer_scaled, columns=df_outliers.drop('ID',axis=1).columns)
df_customer_scaled
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit_transform(df_customer_scaled)
pca_customer_data_scaled = pca.transform(df_customer_scaled)
pca_customer_data_scaled
```

Analizando el ratio de varianza

```python
var = pca.explained_variance_ratio_
print(var) #Con un indice tengo la varianza [0], con 2 indices [1] y así sucesivamente en el array
```

```python
cum_var = np.cumsum(np.round(var,decimals=4)*100) #Acumulado del array redondeando la varianza con 4 decimales
plt.figure(figsize=(10,10))
plt.plot(cum_var,'r-x')
plt.show()
```

```python
pca_customer_standard = pd.DataFrame(pca_customer_data_scaled)
pca_customer_standard.drop([11,12,13,14,15],axis=1,inplace=True) #Elimino las columnas que ya no necesito, inplace = lo deja en nuestro dataframe

```

```python
pca_customer_standard #Tenemos toda la data resumida en 10 componentes principales que no modifica mucho la varianza
```
